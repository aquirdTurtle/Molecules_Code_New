$$
E=k_B T = \frac{\hbar^2 \ell(\ell+1)}{2\mu R^2}
$$

```python
def Renergy(ell,rv):
    return mc.hbar**2*ell*(ell+1)/(2*(mc.Rb87_M/2)*rv**2)
```

```python
ellsp = sp.symbols('ell')
ellVals = []
ellRounded = []
deltaEnergies = []
rvals = np.linspace(1e-9,1000e-9,100)
for rv in rvals:
    Kin = mc.k_B*2000e-6 # 10 uK
    sol = sp.solve(Renergy(ellsp,rv)-Kin)
    ellVals.append(max(sol))
    ellRounded.append(round(max(sol)))
    deltaEnergies.append((Renergy(ellRounded[-1] + 1,rv)-Renergy(ellRounded[-1], rv))*1e-6/mc.hbar)
```

```python
deltap = Renergy(2,25e-9) - Renergy(1,25e-9)
deltam = Renergy(1,25e-9) - Renergy(0,25e-9)
print( (deltap/mc.h)*1e-6, (deltam/mc.h)*1e-6)
```
